```python
import numpy as np
import glob
import imageio
import os
import matplotlib.pyplot as plt
```

```python
dataset = []
path = os.path.dirname(os.getcwd()) + '/data/0.5928/*.npy'
for filename in glob.glob(path):
    with open(os.path.join(os.getcwd(), filename), 'rb') as f: 
        dataset.append(np.load(f).reshape(128,128,1))
```

```python
fig = plt.figure(figsize=(4, 4))
plt.imshow(dataset[0], cmap='gray')
plt.axis('off')
```
